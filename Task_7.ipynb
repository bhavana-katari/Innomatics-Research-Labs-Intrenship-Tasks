Problem 1: Smart Light Controller

Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.

üåç Real-Time Use

Smart home automation systems

üì• Sample Input

Light Name: Bedroom Light

Action: ON

üì§ Expected Output

Bedroom Light is ON


```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")
light_name = "Bedroom Light"
action = "ON"

bedroom_light = SmartLight(light_name)

if action == "ON":
    bedroom_light.turn_on()
elif action == "OFF":
    bedroom_light.turn_off()

bedroom_light.display_status()
```

Problem 2: Employee ID Card System

 Problem Statement

Create a class that stores employee name, ID, and department and displays ID card details.

Real-Time Use

Corporate HR systems

üì• Sample Input

Employee Name: Rahul

Employee ID: EMP102

Department: IT

üì§ Expected Output

Employee ID Card

Name: Rahul

ID: EMP102

Department: IT



```python
class EmployeeIDCard:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"

rahul_card = EmployeeIDCard(employee_name, employee_id, department)

rahul_card.display_card()
```
